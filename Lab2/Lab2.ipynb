### Câu 10: Kiểm tra kiểu dữ liệu của mỗi attribute trong df. Sau đó, chuyển về kiểu dữ liệu cho phù hợp.

```python
data_10 = data_8.copy()
data_10.dtypes
```

```python
data_10['normalized-losses'] = data_10['normalized-losses'].astype('float')
data_10['wheel-base '] = data_10['wheel-base'].astype('float')
data_10['length'] = data_10['length'].astype('float')
data_10['width'] = data_10['width'].astype('float')
data_10['curb-weight'] = data_10['curb-weight'].astype('float')
data_10['bore'] = data_10['bore'].astype('float')
data_10['stroke'] = data_10['stroke'].astype('float')
data_10['horsepower'] = data_10['horsepower'].astype('float')
data_10['peak-rpm'] = data_10['peak-rpm'].astype('float')
data_10['price'] = data_10['price'].astype('float')
```

### Câu 11: Chuẩn hóa "length", "width" và "height" trong phạm vi [0,1].

```python
data_11 = data_10.copy()
```

```python
data_11['length'] = (data_11['length']-data_11['length'].min()) / (data_11['length'].max() - data_11['length'].min())
data_11['height'] = (data_11['height']-data_11['height'].min()) / (data_11['height'].max() - data_11['height'].min())
```

### Câu 12: Tạo các bin là ['Low', 'Medium', 'High'] cho “horsepower”

```python
data_12 = data_11.copy()
bins = np.linspace(min(data_11['horsepower']), max(data_11['horsepower']), 4)
group_names = ['Low', 'Medium', 'High']
data_12['horsepower-binned'] = pd.cut(data_12['horsepower'], bins, labels=group_names, include_lowest = True)
data_12['horsepower-binned']
```

### Câu 13: Viết hàm lưu các df trong quá trình xử lý dữ liệu của các câu trên thành các tập tin sau “.csv”, hoặc “.json”, hoặc “excel”, hoặc “hdf”, hoặc “.sql”. Đặt tên file lưu trữ theo fomat sau: df_cau0x.&lt;kiểu file&gt;, với x là số thứ tự câu.

```python
def cau13(df, file_style, number):
    if file_style == 'csv':
        return df.to_csv('df_cau0'+str(number)+'.csv')
    if file_style == 'json':
        return df.to_json('df_cau0'+str(number)+'.json')
    if file_style == 'excel':
        return df.to_excel('df_cau0'+str(number)+'.xlsx')
    if file_style == 'hdf':
        return df.to_hdf('df_cau0'+str(number)+'.h5') 
    #if file_style == 'sql':
        #return df.to_sql('df_cau0'+str(number)+'.sql')
```
