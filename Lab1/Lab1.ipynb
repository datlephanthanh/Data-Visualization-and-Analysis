### Câu 8: Tìm các sample mà attribute “price” có ký hiệu là “?”, hoặc “.”, hoặc “”

```python
def cau8(attribute, s):
    if s == "?":
        return data[attribute == "?"]
    if s == ".":
        return data[attribute == "."]
    if s == "":
        return data[attribute == ""]
```

```python
cau8(data['price'], '?')
```

### Câu 12: Viết hàm lưu dataframe bất kỳ thành các tập tin sau “.csv”, “.json”, “excel”, “hdf”, “.sql”. Sau đó áp dụng lưu trữ các df đã có (câu 1, 3, 6...). Đặt tên file lưu trữ theo fomat sau: df_cau0x.<kiểu file>, với x là số thứ tự câu.

```python
def cau12(df, file_style, x):
    if file_style == 'csv':
        return df.to_csv('df_cau0'+str(x)+'.csv')
    if file_style == 'json':
        return df.to_json('df_cau0'+str(x)+'.json')
    if file_style == 'excel':
        return df.to_excel('df_cau0'+str(x)+'.xlsx')
    if file_style == 'hdf':
        return df.to_hdf('df_cau0'+str(x)+'.h5') 
    #if file_style == 'sql':
        #return df.to_sql('df_cau0'+str(x)+'.sql')
```

```python
cau12(data, 'csv', 12)
```
